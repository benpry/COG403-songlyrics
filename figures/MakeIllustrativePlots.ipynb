```python
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams["figure.figsize"] = (9, 6)
```

```python
x1 = np.arange(0, 20, 0.2)
y1 = np.sin(x1 / 3)
plt.scatter(x1, y1, label="y1")

x2 = np.arange(0, 20, 0.2)
y2 = np.sin((x2 - 5) / 3)
plt.scatter(x2, y2, label="y2")

plt.legend()
plt.xlabel("time")
plt.ylabel("y")
plt.title("Two Time Series")
plt.ylim(bottom=-2, top=2)
plt.savefig("timelag_1.png", bbox_inches="tight")
```

```python
plt.scatter(y1, y2)
plt.xlabel("y1")
plt.ylabel("y2")
plt.title("Scatterplot of Time Series Data")
plt.savefig("corr1.png", bbox_inches="tight")
```

```python
x1 = np.arange(0, 20, 0.2)
y1 = np.sin(x1 / 3)
plt.scatter(x1, y1, label="y1")

x2 = np.arange(0, 15, 0.2)
y2 = np.sin(x2 / 3)
plt.scatter(x2, y2, label="y2")

plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.title("Two time Series")
plt.ylim(bottom=-2, top=2)
plt.savefig("timelag_2.png", bbox_inches="tight")
```

```python
plt.scatter(y1[:len(y2)], y2)
plt.xlabel("y1")
plt.ylabel("y2")
plt.title("Correlation Between Time Series - offset")
plt.savefig("corr_2.png", bbox_inches="tight")
```
